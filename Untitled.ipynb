```python
import pandas as pd
import seaborn as sns
import os
%matplotlib inline
os.chdir("C://Users/ismail/Desktop")
calls=pd.read_csv("stage.csv")
calls.head()



```

```python
calls.dropna(how='all',inplace=True)
```

```python
calls.fillna(method='ffill',inplace=True)
```

```python
t=[]
x=calls['model']
c=[]
modele={}
l=1
c.append(x[0])
```

```python
for i in range(1,len(x)):
    j=0
    while(j<len(c)):
        if x[i]!=c[j]:
            j+=1
        else:
            break
    if (j==len(c)):
        c.append(x[i])
```

```python
for k in range(0,len(c)):
   modele[c[k]]=l
   l+=1
```

```python
def models(y):
    for i in modele.keys():
        if y==i:
            return modele[i]
calls['modele2']=calls['model'].apply(models)
```

```python
w=[]
y=calls['make']
d=[]
l=1
d.append(y[0])
```

```python
for i in range(1,len(y)):
    j=0
    while(j<len(d)):
        if y[i]!=d[j]:
            j+=1
        else:
            break
    if (j==len(d)):
        d.append(y[i])
```

```python
make_all={}
for k in range(0,len(d)):
   make_all[d[k]]=l
   l+=1
```

```python
def make1(y):
    for i in make_all.keys():
        if y==i:
            return make_all[i]
calls['make2']=calls['make'].apply(make1)
```

```python
def carburant1(x):
    if x=="essence":
        return 1
    if x=="diesel":
        return 0
    if x=="hybride":
        return 2
    if x=="electrique":
        return 3
calls['Carburant_num']=calls['carburant'].apply(carburant1)  
calls
```
